```python
# ============================================
# Install Dependencies (Google Colab)
# ============================================

!pip install -q gradio google-genai pillow

# ============================================
# Imports
# ============================================

import gradio as gr
from google import genai
from PIL import Image
import tempfile
import os

# ============================================
# Gemini API Setup
# ============================================

API_KEY = "YOUR_GEMINI_API_KEY"

client = genai.Client(api_key='AIzaSyBESEZWmbui6UDLi67alurK9O0f4e_g2QQ')

# ============================================
# Medical Doctor Assistant Function
# ============================================

def medical_assistant(
    image,
    symptoms,
    department,
    urgency,
    age_group,
    gender,
    analysis_type,
    response_style,
):

    try:

        if image is None:
            return "❌ Please upload a medical image."

        # Save uploaded image temporarily
        temp_file = tempfile.NamedTemporaryFile(delete=False, suffix=".png")
        image.save(temp_file.name)

        # Upload image to Gemini
        uploaded_file = client.files.upload(file=temp_file.name)

        # Advanced Gemini Prompt
        prompt = f"""
You are an advanced AI Medical Doctor Assistant.

Analyze the uploaded medical image carefully and provide a professional medical-style analysis.

PATIENT DETAILS:
- Symptoms: {symptoms}
- Medical Department: {department}
- Urgency Level: {urgency}
- Age Group: {age_group}
- Gender: {gender}
- Analysis Type: {analysis_type}
- Response Style: {response_style}

INSTRUCTIONS:
1. Analyze the uploaded image carefully.
2. Identify visible abnormalities, conditions, or observations.
3. Provide possible medical explanations.
4. Suggest precautions and next steps.
5. Recommend specialist consultation if necessary.
6. Mention emergency warning signs if visible.
7. Keep tone professional and medically structured.
8. Use markdown headings and bullet points.
9. Include confidence level for observations.
10. DO NOT claim guaranteed diagnosis.

IMPORTANT:
- This is an AI assistant and not a licensed doctor.
- Encourage professional medical consultation.
"""

        # Gemini Response
        response = client.models.generate_content(
            model="gemini-3-flash-preview",
            contents=[uploaded_file, prompt],
        )

        # Remove temp file
        os.unlink(temp_file.name)

        return response.text

    except Exception as e:
        return f"Error: {str(e)}"


# ============================================
# Custom Styling
# ============================================

custom_css = """

body {
    background: linear-gradient(135deg, #0f172a, #1e3a8a);
}

.gradio-container {
    font-family: 'Poppins', sans-serif;
}

.main-title {
    text-align: center;
    font-size: 42px;
    font-weight: bold;
    color: white;
    margin-bottom: 10px;
}

.subtitle {
    text-align: center;
    color: #cbd5e1;
    font-size: 18px;
    margin-bottom: 25px;
}

.glass {
    background: rgba(255,255,255,0.08);
    backdrop-filter: blur(12px);
    border-radius: 18px;
    padding: 15px;
}

"""

# ============================================
# Gradio UI
# ============================================

with gr.Blocks(
    theme=gr.themes.Soft(),
    css=custom_css
) as demo:

    # Header
    gr.HTML("""
    <div class="main-title">
        🩺 AI Medical Doctor Assistant
    </div>

    <div class="subtitle">
        Upload medical images and get AI-powered professional analysis
    </div>
    """)

    with gr.Row():

        # LEFT SIDE
        with gr.Column(scale=1):

            with gr.Group(elem_classes="glass"):

                image_input = gr.Image(
                    type="pil",
                    label="📷 Upload Medical Image"
                )

                symptoms = gr.Textbox(
                    lines=5,
                    placeholder="Describe symptoms, pain, duration, fever, swelling, etc.",
                    label="📝 Patient Symptoms"
                )

        # RIGHT SIDE
        with gr.Column(scale=1):

            with gr.Group(elem_classes="glass"):

                department = gr.Dropdown(
                    [
                        "General Medicine",
                        "Dermatology",
                        "Radiology",
                        "Cardiology",
                        "Orthopedics",
                        "Neurology",
                        "Dentistry",
                        "Ophthalmology",
                        "ENT",
                        "Emergency Care"
                    ],
                    value="General Medicine",
                    label="🏥 Medical Department"
                )

                urgency = gr.Radio(
                    ["Low", "Medium", "High", "Emergency"],
                    value="Medium",
                    label="🚨 Urgency Level"
                )

                age_group = gr.Dropdown(
                    [
                        "Child",
                        "Teen",
                        "Adult",
                        "Senior Citizen"
                    ],
                    value="Adult",
                    label="👤 Age Group"
                )

                gender = gr.Radio(
                    ["Male", "Female", "Other"],
                    value="Male",
                    label="⚧ Gender"
                )

                analysis_type = gr.Dropdown(
                    [
                        "Basic Analysis",
                        "Detailed Analysis",
                        "Risk Assessment",
                        "Treatment Suggestions",
                        "Full Medical Report"
                    ],
                    value="Detailed Analysis",
                    label="🧠 Analysis Type"
                )

                response_style = gr.Dropdown(
                    [
                        "Professional",
                        "Simple Explanation",
                        "Detailed Clinical",
                        "Short Summary"
                    ],
                    value="Professional",
                    label="📋 Response Style"
                )

    # Generate Button
    generate_btn = gr.Button(
        "🔍 Analyze Medical Image",
        variant="primary"
    )

    # Output
    output = gr.Markdown(
        label="📄 AI Medical Report"
    )

    # Button Click
    generate_btn.click(
        fn=medical_assistant,
        inputs=[
            image_input,
            symptoms,
            department,
            urgency,
            age_group,
            gender,
            analysis_type,
            response_style,
        ],
        outputs=output
    )

# ============================================
# Launch App
# ============================================

demo.launch(debug=True)
```
